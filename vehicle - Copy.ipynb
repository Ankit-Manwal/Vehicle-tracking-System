```python
from roboflow import Roboflow
rf = Roboflow(api_key="unauthorized")
project = rf.workspace("roboflow-universe-projects").project("license-plate-recognition-rxg4e")
version = project.version(13)
dataset = version.download("yolov8")
                
```

```python
from ultralytics import YOLO
import shutil
import os
```

```python
# -----------------------------
# Load pretrained YOLOv8 nano
# -----------------------------
model = YOLO("yolov8n.pt")  # auto-downloads if not present

# -----------------------------
# Train the model
# -----------------------------
results = model.train(
    data="data.yaml",     # path to your data.yaml
    epochs=5,
    imgsz=640,
    batch=16,
    workers=2,
    device=0               # set to "cpu" if no GPU
)



# -----------------------------
# Save trained weights
# -----------------------------
os.makedirs("saved_models", exist_ok=True)

best_weights = "runs/detect/train/weights/best.pt"
last_weights = "runs/detect/train/weights/last.pt"

if os.path.exists(best_weights):
    shutil.copy(best_weights, "saved_models/license_plate_best.pt")

if os.path.exists(last_weights):
    shutil.copy(last_weights, "saved_models/license_plate_last.pt")

print("âœ… Training complete. Models saved in /saved_models")
```
